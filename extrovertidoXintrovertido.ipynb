```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
import kagglehub
import shutil
import os

# Baixa o dataset e retorna o caminho temporário onde ele foi salvo
temp_path = kagglehub.dataset_download("rakeshkapilavai/extrovert-vs-introvert-behavior-data")

# Define o caminho final em /content
dest_path = "/content/extrovert-vs-introvert-behavior-data"

# Cria o diretório se ainda não existir
os.makedirs(dest_path, exist_ok=True)

# Move os arquivos do caminho temporário para /content
shutil.copytree(temp_path, dest_path, dirs_exist_ok=True)

print("Dataset copiado para:", dest_path)
```

```python
df1 = pd.read_csv("/content/extrovert-vs-introvert-behavior-data/personality_datasert.csv")
df2 = pd.read_csv("/content/extrovert-vs-introvert-behavior-data/personality_dataset.csv")
```

```python
df1
```

```python
df2.columns
```

```python
one_hot = ['Stage_fear', 'Drained_after_socializing']
```

```python
for col in df2.columns:
    print(df2[col].unique())
```

```python
df_one_hot = pd.get_dummies(df2,columns=one_hot)
```

```python
df_one_hot.dropna(inplace=True)
```

```python
df_one_hot['Personality'] = df_one_hot['Personality'].map({'Introvert': 0, 'Extrovert': 1})
```

```python
df_one_hot
```

```python
X = df_one_hot.drop('Personality', axis=1)
y = df_one_hot['Personality']
```

```python

normalizador = sk.preprocessing.MinMaxScaler()

X_norm = normalizador.fit_transform(X)
```

```python
import joblib

joblib.dump(normalizador, 'normalizador_scaler.pkl')
```

```python
X_treino, X_teste, y_treino, y_teste = sk.model_selection.train_test_split(X_norm, y, test_size=0.2, random_state=42)
```

```python
pd.DataFrame(X_treino)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(X_treino.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_treino, y_treino, epochs=20, batch_size=32, validation_split=0.2)
```

```python
model.evaluate(X_teste, y_teste)
```
